```python
def requrl(url):
  up = urlparse(url)
  nl = up.netloc
  query = up.query
  path = up.path
  params = up.params
  fragment = up.fragment


  try:
    res = get_tld(url, as_object=True)

    tld = res.tld
  except:
    print("No res")
    tld = " "

  # print(tld)
  urlparts = nl.rsplit(tld)
  urlpart = urlparts[0]

  required_url = urlpart+query+path+params+fragment

  # print("Final Url: ", required_url)
  return required_url
```

```python
def character_embedder(url):
    url = url.lower()
    char_arr = []
    char_arr.extend(url)
    tokens_count = Counter(url)
    url_vector = {}
    for i in char_arr:
      token_count = tokens_count[i]
      key = i
      value = token_count
      url_vector[key] = value
    return url_vector
```

```python
def normalize(arr, t_min, t_max):
    norm_arr = []
    diff = t_max - t_min
    diff_arr = max(arr) - min(arr)
    if diff_arr == 0:
      for i in arr:
        temp = 1
        norm_arr.append(temp)
    else:
      for i in arr:
        temp = (((i - min(arr))*diff)/diff_arr) + t_min
        norm_arr.append(temp)

    return norm_arr
```

```python
def urlVector(url_vector):
  dictionary = "abcdefghijklmnopqrstuvwxyz0123456789!#$&|()*+,/:;=?@[]-_.~"
  url_vec = []
  for i in dictionary:
    if i in url_vector.keys():
      url_vec.append(url_vector.get(i))
    else:
      url_vec.append(0)
  return url_vec
```
